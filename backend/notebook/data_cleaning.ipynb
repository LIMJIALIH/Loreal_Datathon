```python
import re

def iso8601_to_seconds(duration_str):
    if pd.isna(duration_str):
        return None

    # Handle 'P' and 'T' separation
    if 'D' in duration_str:
        parts = duration_str.split('T')
        day_part = parts[0]
        time_part = parts[1] if len(parts) > 1 else ''
    else:
        day_part = ''
        time_part = duration_str.replace('PT', '')

    total_seconds = 0

    # Parse days
    day_match = re.search(r'(\d+)D', day_part)
    if day_match:
        total_seconds += int(day_match.group(1)) * 86400

    # Parse hours, minutes, seconds from time part
    hour_match = re.search(r'(\d+)H', time_part)
    if hour_match:
        total_seconds += int(hour_match.group(1)) * 3600

    minute_match = re.search(r'(\d+)M', time_part)
    if minute_match:
        total_seconds += int(minute_match.group(1)) * 60

    second_match = re.search(r'(\d+)S', time_part)
    if second_match:
        total_seconds += int(second_match.group(1))

    return total_seconds

df['duration_seconds'] = df['contentDuration'].apply(iso8601_to_seconds)
display(df[['contentDuration', 'duration_seconds']].head())
```

```python
# The tags column is already cleaned with empty strings for NaN values.
# We need to handle cases where tags might be represented as strings that look like lists, e.g., "['tag1', 'tag2']"
# or just single strings.
# Assuming the tags are in a string format that looks like a list or a single tag string,
# we can try to parse them or count based on common separators.

def count_tags(tag_list_str):
    if isinstance(tag_list_str, str) and tag_list_str.strip() != '':
        # Attempt to parse as a list string, or split by common separators if not a list string
        try:
            # Safely evaluate the string if it looks like a list
            if tag_list_str.strip().startswith('[') and tag_list_str.strip().endswith(']'):
                 tags = eval(tag_list_str)
                 if isinstance(tags, list):
                     return len(tags)
            # Otherwise, assume space or comma separation and split
            return len([tag for tag in re.split(r'[,\s]+', tag_list_str) if tag])
        except:
            # If eval fails or splitting doesn't work as expected, return 1 for a non-empty string
            return 1
    return 0

df['num_tags'] = df['tags'].apply(count_tags)
display(df[['tags', 'num_tags']].head())
```

```python
import re

def extract_topic_name(topic_list_str):
    if pd.isna(topic_list_str):
        return None

    # Safely evaluate the string if it looks like a list
    try:
        if topic_list_str.strip().startswith('[') and topic_list_str.strip().endswith(']'):
            topics = eval(topic_list_str)
            if isinstance(topics, list):
                extracted_topics = []
                for topic_url in topics:
                    match = re.search(r'/wiki/(.+)', topic_url)
                    if match:
                        extracted_topics.append(match.group(1).replace('_', ' '))
                return ', '.join(extracted_topics)
    except:
        pass # If eval fails, continue to the next part

    # If not a list string or eval failed, check if it's a single URL string
    if isinstance(topic_list_str, str):
        match = re.search(r'/wiki/(.+)', topic_list_str)
        if match:
            return match.group(1).replace('_', ' ')

    return None

df['topicCategories'] = df['topicCategories'].apply(extract_topic_name)
display(df['topicCategories'].head())
```
