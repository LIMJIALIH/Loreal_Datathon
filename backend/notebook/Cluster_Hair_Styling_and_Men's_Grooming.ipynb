## Embed text

### Subtask:
Generate embeddings for the `clean_text` column.


**Reasoning**:
Generate embeddings for the `clean_text` column using a pre-trained sentence transformer model and store them in a new variable.



```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
text_embeddings = model.encode(df['clean_text'].tolist())
```

## Cluster embeddings

### Subtask:
Cluster the generated embeddings into a maximum of 50 clusters.


**Reasoning**:
Cluster the generated embeddings using KMeans as instructed.



```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=50, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(text_embeddings)
```

## Assign keywords to clusters

### Subtask:
Assign a descriptive keyword to each cluster.


**Reasoning**:
Assign the cluster labels to a new column in the DataFrame and then group the DataFrame by this new column to prepare for keyword extraction per cluster.



```python
df['cluster_label'] = cluster_labels
grouped_clusters = df.groupby('cluster_label')
```

**Reasoning**:
Iterate through each cluster group, extract the text, calculate word frequencies within each cluster, and identify the most frequent word or phrase as the keyword for that cluster.



**Reasoning**:
The error indicates that `filtered_word_counts` is a dictionary and does not have a `most_common` method. This is because the result of the dictionary comprehension is a dictionary, not a Counter object. The code needs to apply the filtering to the Counter object directly before finding the most common word.



```python
from collections import Counter
import re

cluster_keywords = {}

for cluster_id, group in grouped_clusters:
    # Convert all items to string before joining
    all_text_in_cluster = " ".join(group['clean_text'].astype(str).tolist())
    words = re.findall(r'\b\w+\b', all_text_in_cluster.lower())
    word_counts = Counter(words)
    # Exclude common English stop words and single characters
    stop_words = set(['the', 'and', 'a', 'to', 'of', 'is', 'in', 'it', 'you', 'for', 'on', 'with', 'that', 'this', 'get', 'how', 'your', 'men', 'tips', 'hair', 'style'])
    # Apply filtering to the Counter object
    filtered_word_counts = Counter({word: count for word, count in word_counts.items() if word not in stop_words and len(word) > 1})

    if filtered_word_counts:
        # Find the most common word as the keyword
        most_common_word = filtered_word_counts.most_common(1)[0][0]
        cluster_keywords[cluster_id] = most_common_word
    else:
        cluster_keywords[cluster_id] = "unknown"

print("Cluster Keywords:")
for cluster_id, keyword in cluster_keywords.items():
    print(f"Cluster {cluster_id}: {keyword}")
```

## Map videos to clusters

### Subtask:
Map each video entry to its closest cluster and assign the corresponding keyword.


**Reasoning**:
Map the cluster labels to the corresponding keywords and add them as a new column to the dataframe.



```python
df['keyword'] = df['cluster_label'].map(cluster_keywords)
display(df[['title', 'cluster_label', 'keyword']].head())
```

```python
import pandas as pd
import ast


def safe_string_to_list(s):
    try:
        return ast.literal_eval(s) if isinstance(s, str) and s.startswith('[') and s.endswith(']') else [s]
    except (ValueError, SyntaxError):
        return [s] if isinstance(s, str) else []

df['keywords'] = df['keywords'].apply(safe_string_to_list)

display(df.head())
```

```python
from sentence_transformers import SentenceTransformer

unique_keywords = df['keywords'].explode().unique().tolist()

model = SentenceTransformer('all-MiniLM-L6-v2')

keyword_embeddings = model.encode(unique_keywords)

keyword_embedding_df = pd.DataFrame({'keyword': unique_keywords, 'embedding': list(keyword_embeddings)})
display(keyword_embedding_df.head())
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=25, random_state=42, n_init=10)
kmeans.fit(list(keyword_embedding_df['embedding']))

keyword_embedding_df['cluster_label'] = kmeans.labels_
display(keyword_embedding_df.head())
```

```python
from sklearn.metrics.pairwise import euclidean_distances

# Calculate cluster centroids
cluster_centroids = kmeans.cluster_centers_

# Calculate distances from each keyword to its cluster centroid
distances = euclidean_distances(list(keyword_embedding_df['embedding']), cluster_centroids)

# Find the index of the representative keyword for each cluster
representative_keyword_indices = {}
for cluster_label in range(kmeans.n_clusters):
    cluster_indices = keyword_embedding_df[keyword_embedding_df['cluster_label'] == cluster_label].index
    distances_to_centroid = distances[cluster_indices, cluster_label]
    representative_index_in_cluster = cluster_indices[distances_to_centroid.argmin()]
    representative_keyword_indices[cluster_label] = representative_index_in_cluster

# Create a mapping from cluster label to representative keyword
cluster_to_representative_keyword = {
    cluster_label: keyword_embedding_df.loc[index, 'keyword']
    for cluster_label, index in representative_keyword_indices.items()
}

# Create a mapping from all keywords to their representative keywords
keyword_to_cluster = keyword_embedding_df.set_index('keyword')['cluster_label'].to_dict()
keyword_to_representative = {
    keyword: cluster_to_representative_keyword[cluster_label]
    for keyword, cluster_label in keyword_to_cluster.items()
}

# Store the mapping
representative_keywords = cluster_to_representative_keyword

display(representative_keywords)
```

```python
# Replace keywords in the DataFrame with representative keywords
df['new_keywords'] = df['keywords'].apply(lambda keyword_list: [
    keyword_to_representative.get(keyword, keyword) for keyword in keyword_list
])

# Display the head of the updated df to show the new column
display(df.head())
```

```python
# Load the cleaned_videos.csv file
cleaned_videos_df = pd.read_csv("/content/cleaned_videos.csv", engine='python')

# Select only the videoId and publishedAt columns
cleaned_videos_subset = cleaned_videos_df[['videoId', 'publishedAt']]

# Merge the dataframes on videoId
df = df.merge(cleaned_videos_subset, on='videoId', how='left')

# Display the head of the updated df to show the new column
display(df.head(20))
```

```python
# === Simplified Keyword Trend Detection ===
import pandas as pd
import numpy as np

# Step 1: Preprocess
df["publishedAt"] = pd.to_datetime(df["publishedAt"], errors="coerce")
df = df.dropna(subset=["keywords"]).sort_values("publishedAt")
df["keyword"] = df["keywords"].str.strip().str.lower()
df["month"] = df["publishedAt"].dt.to_period("M")

# Step 2: Aggregate mentions per month + engagement
keyword_time = (
    df.groupby(["keyword", "month"])
      .agg({"videoId":"count", "engagement_rate":"mean"})
      .reset_index()
      .rename(columns={"videoId":"mentions"})
)

# Step 3: Compute recent velocity (last 3 months slope)
def recent_velocity(x):
    if len(x) < 2:
        return 0
    # take last 3 months
    t = np.arange(len(x))[-3:]
    y = x["mentions"].values[-3:]
    return np.polyfit(t, y, 1)[0]

velocities = keyword_time.groupby("keyword").apply(recent_velocity).reset_index(name="velocity")

# Step 4: Average engagement per keyword
eng = keyword_time.groupby("keyword")["engagement_rate"].mean().reset_index()

# Step 5: Merge velocity + engagement
score_df = velocities.merge(eng, on="keyword", how="left")

# Step 6: Classification thresholds
eng_thresh = score_df["engagement_rate"].median()

def classify_phase(velocity, engagement, v_thresh=0, e_thresh=eng_thresh):
    if velocity > v_thresh and engagement < e_thresh:
        return "Emerging"   # upward mentions, low engagement
    elif velocity > v_thresh and engagement >= e_thresh:
        return "Growing"    # upward mentions, high engagement
    elif velocity <= v_thresh and engagement >= e_thresh:
        return "Peaking"    # slowing mentions, still high engagement
    else:
        return "Decaying"   # losing both mentions and engagement

score_df["phase"] = score_df.apply(
    lambda r: classify_phase(r["velocity"], r["engagement_rate"]), axis=1
)

# Step 7: Ranking logic for Top 10
# - Emerging & Growing → rank by velocity (higher = better momentum)
# - Peaking → rank by engagement_rate (since velocity flat/negative but still resonates)
# - Decaying → rank by velocity (more negative = faster decay)

print("🔝 Top Keywords by Phase\n" + "-"*60)
for phase in ["Emerging", "Growing", "Peaking", "Decaying"]:
    df_phase = score_df[score_df["phase"] == phase].copy()

    if phase in ["Emerging", "Growing"]:
        df_phase = df_phase.sort_values("velocity", ascending=False)
    elif phase == "Peaking":
        df_phase = df_phase.sort_values("engagement_rate", ascending=False)
    else:  # Decaying
        df_phase = df_phase.sort_values("velocity", ascending=True)  # more negative first

    print(f"\n🔝 Top 10 {phase} keywords:")
    print(df_phase.head(10)[["keyword", "velocity", "engagement_rate", "phase"]].to_string(index=False))
```

```python
# === Step 7: Save to CSV ===
score_df.to_csv("Hair_Styling_and_Men's_Grooming_keyword_trend_phases.csv", index=False)
print("✅ Saved to Hair_Styling_and_Men's_Grooming_keyword_trend_phases.csv")
```
